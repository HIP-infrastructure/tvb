# Welcome to the TVB HIP app

This is the Virtual Brain (TVB) HIP app, which provides access to the TVB simulation tools, as well as integrations with the wider EBRAINS ecosystem.

## Next steps

Please look through the following links depending on your needs. 

- Reconstruction: building a virtual brain model from raw data
- Simulation: run simulations of brain model
- Inference: compute model parameters from functional data
- EBRAINS integrations
  - KG & Atlas services: receive data from EBRAINS sources
  - HPC services: run computationally intensive models on HPC resources
